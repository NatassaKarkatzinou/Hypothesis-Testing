# I. Introduction and core concepts
<br>

<a id="what-is-hypothesis-testing"></a>
## 1. What is Hypothesis Testing

Hypothesis testing is a **statistical decision process** used to determine whether there is enough evidence in a **sample of data** to infer that a certain condition is true for the **entire population**.

It compares what we **observe in data** with what we would **expect to observe if nothing interesting happened** (the *null hypothesis*).

Typical workflow:
1. Formulate hypotheses
2. Choose a significance level $\alpha$
3. Compute a test statistic
4. Determine the p-value or rejection region
5. Make a decision (reject or not reject $H_0$)  

<a id="null-and-alternative-hypotheses"></a>
## 2. Null and Alternative Hypotheses

- The **null hypothesis** $H_0$ represents the default assumption or "status quo".
- The **alternative hypothesis** $H_1$ represents the claim we want to test.

Example:
- $H_0$: The mean lifetime of a battery is 1000 hours.
- $H_1$: The mean lifetime is different from 1000 hours.

We then collect sample data and use it to decide whether to reject $H_0$.

<a id="type-i-and-type-ii-errors"></a>
## 3. Type I and Type II Errors

There are two possible kinds of error in hypothesis testing:

| True situation | Decision: Reject $H_0$ | Decision: Do not reject $H_0$ |
|----------------|-------------------------|--------------------------------|
| $H_0$ is true  | ❌ Type I error (false positive, probability $\alpha$) | ✅ Correct decision |
| $H_0$ is false | ✅ Correct decision | ❌ Type II error (false negative, probability $\beta$) |

The **significance level** $\alpha$ (e.g., 0.05) is the probability of making a Type I error.

The **power** of a test is $1 - \beta$ and denotes the probability of correctly rejecting $H_0$ when it is false.

<a id="p-values-and-critical-regions"></a>
## 4. p-values and Critical Regions

- The **p-value** is the probability of obtaining a test statistic as extreme as the observed one, assuming that $H_0$ is true.

- A **small p-value** (typically < 0.05) suggests that the observed data is unlikely under $H_0$, and we may **reject** the null hypothesis.

- The **critical region** is the range of values for which we reject $H_0$.

Example for a two-tailed test:
$$
|T| > t_{1-\alpha/2, \, df}
$$


Visually, the sampling distribution of the test statistic under $H_0$ has most of its mass near the center.  
Extreme values (the tails) are unlikely if $H_0$ is true. These form the **critical regions**.

The **p-value** is the total area of the tails beyond the observed statistic.  
**Reject $H_0$ if the observed statistic falls in the shaded region or if the p-value ≤ α.**

<a id="one-tailed-vs-two-tailed-tests"></a>
## 5. One-Tailed vs Two-Tailed Tests

Depending on the research question, the alternative hypothesis may be:

- **Two-tailed:**     $\ \  \quad H_0:\ \mu = \mu_0, \qquad H_1:\ \mu \neq \mu_0$

- **Right-tailed:**  $ \quad H_0:\ \mu = \mu_0, \qquad H_1:\ \mu > \mu_0$

- **Left-tailed:**    $\ \  \quad H_0:\ \mu = \mu_0, \qquad H_1:\ \mu < \mu_0$

The tail determines which part of the sampling distribution counts as "extreme" (the rejection region).


<a id="parametric-vs-nonparametric-tests"></a>
## 6. Parametric vs Nonparametric Tests

| Type | Assumptions | Examples |
|------|--------------|-----------|
| **Parametric** | Data come from a population following a known distribution (usually normal). | z-test, t-test, ANOVA |
| **Nonparametric** | No specific distribution is assumed. Works with ranks or medians. | Wilcoxon, Mann–Whitney, Kruskal–Wallis |

Nonparametric tests are useful when data are skewed, ordinal, or when sample sizes are small.

# II. One Sample Tests  
  
In this category of tests, we have 
- a **population** with a **true unknown mean $\mu$**
- a **sample** of size **$n$** with an **sample mean $\bar x$**
- we want to test whether this sample suggests that the population mean $\mu$ is different from a known or claimed value $\mu_0$  
> eg. a manufacturer claims that batteries last 100 hours ($\mu_0$) on average. We test 50 batteries ($n$) and find a sample mean of 103 hours ($\bar x$).
Are the batteries truly better, or could this difference be due to sampling randomness?  

<br>

<a id="one-sample-z-test-known-population-variance-n-large"></a>
## 7. One-Sample z-Test (Known Population Variance $\sigma^2$ or $n \geq 30$)

We use the one-sample z-test when:

- The **population standard deviation $\sigma$** is **known**.
- The **population standard deviation $\sigma$** is **unknown**, but the **sample size is $n \geq 30$**: for large sample sizes, the z-distribution approximates the t-distribution (that is supposed to be used when $\sigma$ is unknown), and it is easier to use. In this case we estimate $\sigma$ by the **sample standard deviation $s$**.  
- The sample is random and follows (approximately) a normal distribution.

### Hypotheses:
$$
H_0: \mu = \mu_0 
\quad\text{vs}\quad
H_1: \mu \neq \mu_0
$$
  
Since the sample is random and follows approximately a normal distribution, then the standardized z-score follows the $\mathcal{N}(0,1)$:  
  
$$
\bar X \sim \mathcal{N}(\mu,\sigma^2/n) \quad \rightarrow \quad Z \sim \mathcal{N}(0,1)
$$  

### Test Statistic
The **z-test statistic:**  
$$
\begin{align}
z_c = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}},
\quad
\quad 
& \text{if } |z_c| > z_{1-\alpha/2} &\longrightarrow \text{Reject the } H_0 \\
& \text{if } p_{value} < \alpha &\longrightarrow \text{Reject the } H_0
\end{align}
$$

tells us how many standard errors away the observed mean is from the null mean.  

- we compare ∣z∣ with the critical value $z_{1-\alpha/2}$:  
  If $ |z_c| <  z_{1-\alpha/2}$, → the difference could easily be random → fail to reject $H_0$.  
  If $ |z_c| >  z_{1-\alpha/2}$, → the difference is unlikely under $H_0$ → reject $H_0$.
- or we calculate the $p-value$ for the corresponding $z_c$:  
  If $p-value < \alpha$, → reject $H_0$.  
  
<br>

In practice, **σ is rarely known**, except in situations like:
- **Quality-control or manufacturing**, where σ is established from huge datasets.
- Simulated or theoretical exercises (teaching, textbooks, etc.).  
  
So, the z-test is mainly used when **we trust σ**, or as an **approximation for large n**.  

<br>

Useful to have in mind:  

<div style="float:left;">

| Confidence Level | z-score |
|------------------|---------|
| 90% | 1.645 |
| 95% | 1.96 |
| 98% | 2.33 |
| 99% | 2.575 |

## 8. One-Sample t-Test (Unknown Population Variance and $n < 30$)

Used when:
- **population standard deviation $\sigma$ is unknown**, and we estimate it using the **sample standard deviation $s$**.
- and the sample size is small: $ n < 30$.

**t-test statistic:**
$$
\begin{align}
t_c = \frac{\bar{x} - \mu_0}{s / \sqrt{n}},
\quad
\quad 
& \text{if } |t_c| > t_{1-\alpha/2} &\longrightarrow \text{Reject the } H_0 \\
& \text{if } p_{value} < \alpha &\longrightarrow \text{Reject the } H_0
\end{align}
$$

This follows a **t-distribution** with $n - 1$ degrees of freedom under $H_0$. 


## 9. One-Sample Proportion Test

### Conditions:
- **categorical (binary) data**: each observation results in one of two outcomes (e.g. success/failure, yes/no, 1/0). 
- **randomness** of the sample 
- **independence** of observations  
- The sampling distribution of $\hat p$ must be approximately **Normal**. For that, it must be: 
  
$$
np_0 \geq 5 \quad and \quad n(1-p_0) \geq 5
$$  
  
  
- The population should be at least 10 times the sample size: $ N \geq 10n$.  
  
*Note: If any of these requirements is violated, we cannot approximate the normal distribution. We can use the exact binomial test instead.*  

  
We test if a **proportion $p$** equals a hypothesized value $p_0$.

### Hypotheses:
$$
H_0: p = p_0
\quad\text{vs}\quad
H_1: p \neq p_0
$$

### Test Statistic
The **z-test statistic:**
$$
\begin{align}
z_c = \frac{\hat{p} - p_0}{\sqrt {\frac{p_0(1 - p_0)}{n}}}, 
\quad
\quad 
& \text{if } |z_c| > z_{1-\alpha/2} &\longrightarrow \text{Reject the } H_0 \\
& \text{if } p_{value} < \alpha &\longrightarrow \text{Reject the } H_0
\end{align}
$$

## 10. Two Independent Samples - Difference of Two Means (Student & Welch)

<br> 

We compare **two *independent* groups** with **numeric outcomes**.

### Assumptions:
- Independence; roughly normal within groups (or large $n$)
- Student’s version adds equal variances.

<br>

### Data Structure
Let samples be
- Group A: $X_1,\dots,X_{n_A}$ with mean $\bar X_A$ and variance $s_A^2$
- Group B: $Y_1,\dots,Y_{n_B}$ with mean $\bar X_B$ and variance $s_B^2$

<br>

### Hypotheses (two-tailed):
$$
H_0:\ \mu_A - \mu_B = 0
\quad\text{vs}\quad
H_1:\ \mu_A - \mu_B \neq 0
$$

- If $\ n_a, n_B\geq30 \quad \longrightarrow \quad \text{z distribution} $  
- If $\ n_A, n_B <30 \quad \longrightarrow \quad \text{t distribution} $  

<br> 

### Test Statistics
**Student’s t-test (equal population variances $\sigma_A^2 = \sigma_B^2 = \sigma^2$):**
Pooled variance (weighted average) and t-statistic  

$$
s_p^2=\frac{(n_A-1)s_A^2+(n_B-1)s_B^2}{n_A+n_B-2},
\qquad
t=\frac{\bar X_A-\bar X_B}{s_p\sqrt{\frac{1}{n_A}+\frac{1}{n_B}}}
$$
with $\text{df}=n_A+n_B-2$.

**Welch’s t-test (unequal population variances):**
$$
t=\frac{\bar X_A-\bar X_B}{\sqrt{\frac{s_A^2}{n_A}+\frac{s_B^2}{n_B}}},
\quad
\text{df}\approx
\frac{\left(\frac{s_A^2}{n_A}+\frac{s_B^2}{n_B}\right)^2}
{\frac{\left(\frac{s_A^2}{n_A} \right)^2}{n_A -1} + \frac{\left(\frac{s_B^2}{n_B} \right)^2}{n_B -1}}
$$

**One-tailed versions:** Replace $H_1$ by $\mu_A>\mu_B$ (right-tailed) or $\mu_A<\mu_B$ (left-tailed); use one-tailed p-values.

<br>

### Effect size
**Cohen’s $d$**  
When we perform a hypothesis test, we are trying to find out whether there is a difference or relationship between two things, e.g. the mean of group A vs. group B. That difference is what we call the effect (or the treatment effect, impact, or effect size). So, the effect is the true magnitude of the phenomenon we are studying.  
  
Hypothesis tests tell us if the effect is real (i.e., unlikely to be due to random chance), but statistical significance doesn’t tell us **how big or important that effect is**. That’s where effect size measures like Cohen’s d come in.

Cohen’s $d$ quantifies how large the difference between two group means is, in units of standard deviation.:  
  
$$ d = \frac{\bar X_A-\bar X_B}{s_p}$$  

$d = 0,2 \rightarrow \text{low effect}$  
$d = 0,5 \rightarrow \text{medium effect}$  
$d = 0,8 \rightarrow \text{large effect}$


```python
import numpy as np
from scipy import stats

np.random.seed(0)
nA, nB = 60, 55
A = np.random.normal(loc=0.0, scale=1.0, size=nA)
B = np.random.normal(loc=0.4, scale=1.3, size=nB)

# Equal-variance check (Levene)
lev_stat, lev_p = stats.levene(A, B)

# Student t-test (assume equal var) and Welch (robust to unequal var)
t_student, p_student = stats.ttest_ind(A, B, equal_var=True)
t_welch,  p_welch  = stats.ttest_ind(A, B, equal_var=False)

# One-tailed p-value example (A > B) from Welch two-sided result:
p_right = p_welch/2 if t_welch > 0 else 1 - p_welch/2

# Cohen's d (pooled)
sp2 = ((nA-1)*A.var(ddof=1) + (nB-1)*B.var(ddof=1)) / (nA+nB-2)
cohens_d = (A.mean() - B.mean()) / np.sqrt(sp2)

print(f"Levene test p={lev_p:.4f} (small p → variances differ)")
print(f"Student t: t={t_student:.3f}, p={p_student:.4f}")
print(f"Welch   t: t={t_welch:.3f},  p={p_welch:.4f}  | one-tailed (A>B) p={p_right:.4f}")
print(f"Cohen's d = {cohens_d:.3f}")
```

```python
# Example  -  Effect Sizes

from scipy.stats import norm
import matplotlib.pyplot as plt

effect_sizes = [0.2, 0.5, 0.8]   # small, medium, large
labels = ["Small (d=0.2)", "Medium (d=0.5)", "Large (d=0.8)"]

x = np.linspace(-4, 6, 600)
mu1, sigma = 0, 1

fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

for ax, d, label in zip(axes, effect_sizes, labels):
    mu2 = mu1 + d          # shift by Cohen's d
    y1 = norm.pdf(x, mu1, sigma)
    y2 = norm.pdf(x, mu2, sigma)
    
    ax.plot(x, y1, color="steelblue", lw=2, label="Group A")
    ax.plot(x, y2, color="tomato", lw=2, label="Group B")
    ax.fill_between(x, np.minimum(y1, y2), color="lightgray", alpha=0.5, label="Overlap")
    
    ax.set_title(label)
    ax.set_xlabel("Value")
    ax.set_xlim(-3, 5)
    ax.legend(fontsize=8)
    ax.grid(alpha=0.2)

axes[0].set_ylabel("Density")
plt.suptitle("Cohen's d — Visualizing Effect Size", fontsize=14, y=1.05)
plt.tight_layout()
plt.show()
```

## 11. Two Proportions Test - A/B Test

We compare the **proportions of "successes"** in **two independent groups** (e.g., conversion rate of version A vs version B).  
  
### Assumptions:
- **categorical (binary) data**: each observation results in one of two outcomes (e.g. success/failure, yes/no, 1/0)
- **randomness** of the sample 
- **independence** of observations  
- The sampling distribution of $\hat p$ must be approximately **Normal**. For that, it must be: 
  
$$
n_1\hat p_1 \ge 5,\ n_1(1-\hat p_1)\ge5,\\
   n_2\hat p_2 \ge 5,\ n_2(1-\hat p_2)\ge5
$$  
  
  
- Population size ≥ 10 × sample size (independence)  

If sample sizes are small or the conditions fail, use **Fisher’s exact test**.

### Hypotheses
For a two-tailed test:
$$
H_0:\ p_A = p_B 
\quad\text{vs}\quad
H_1:\ p_A \ne p_B
$$

For a right-tailed test (Group A higher): $H_1:\ p_A > p_B$

### Test statistic
If the null hypothesis is true (both groups have the same true proportion $p$), then the two sample proportions should be close to each other, apart from random noise. That random variation follows a normal pattern (for large samples), so we can compute:  
  
- the **pooled proportion**  and the  **standard error of the difference**

$$
\hat p = \frac{x_A + x_B}{n_A + n_B}, \quad SE = \sqrt{\hat p (1-\hat p)\!\left(\frac{1}{n_A}+\frac{1}{n_B}\right)}  
$$

- the **z-statistic**
$$
\begin{align}
z_c = \frac{\hat p_A - \hat p_B}{SE},
\quad
\quad 
& \text{if } |z_c| > z_{1-\alpha/2} &\longrightarrow \text{Reject the } H_0 \\
& \text{if } p_{value} < \alpha &\longrightarrow \text{Reject the } H_0
\end{align}
$$

<br>

### Effect size
A useful effect-size measure for proportions is the **risk difference**
$ \hat p_1 - \hat p_2 $  
or **Cohen’s h**
$$
h = 2\!\left[\arcsin\!\sqrt{p_1} - \arcsin\!\sqrt{p_2}\right]
$$
interpreted roughly as: 
small  ≈ 0.2, medium  ≈ 0.5, large  ≈ 0.8.  


```python
from statsmodels.stats.proportion import proportions_ztest, proportion_effectsize

# Example: Website A vs B conversion rates

n1, n2 = 2000, 2100         # samples sizes
x1, x2 = 200, 273           # number of conversions
alpha = 0.05

# z-test for two proportions
stat, pval = proportions_ztest([x1, x2], [n1, n2], alternative="two-sided")

# Effect size (Cohen's h)
h = proportion_effectsize(x1/n1, x2/n2)

print(f"Group A: {x1}/{n1} = {x1/n1:.3f}")
print(f"Group B: {x2}/{n2} = {x2/n2:.3f}")
print(f"z = {stat:.3f}, p = {pval:.4f}")
print(f"Cohen's h = {h:.3f}  →  {(abs(h)<0.2 and 'small' or abs(h)<0.5 and 'medium' or 'large')} effect")

if pval < alpha:
    print("Reject H0: significant difference between proportions.")
else:
    print("Do not reject H0: difference likely due to chance.")
```

The two-proportion z-test compares two estimated proportions and asks:

> "If both groups truly have the same success rate, how likely is it
> to see a difference this large or larger by random chance?"

It is the classical **A/B test**:
- If p-value < α → statistically significant difference.
- If not, observed gap can be due to sampling noise.

Always report both:
- **Statistical significance (p)** → is there a difference?
- **Effect size (Cohen’s h or risk difference)** → how large is it?


```python
import numpy as np
import matplotlib.pyplot as plt

props = np.array([x1/n1, x2/n2])
errs = np.sqrt(props*(1-props)/np.array([n1,n2]))
labels = ["A", "B"]

plt.bar(labels, props, yerr=errs, capsize=6, color=["skyblue","lightcoral"])
plt.title("Two-Sample Proportion Comparison (A/B test)")
plt.ylabel("Proportion of success")
plt.ylim(0, max(props)+0.05)
plt.show()
```

```python
from scipy.stats import norm

# Parameters
n1, n2 = 2000, 2100
p1_H0 = p2_H0 = 0.12          # same proportion under H0
p1_H1, p2_H1 = 0.12, 0.16     # different under H1
alpha = 0.05

# Standard errors
se_H0 = np.sqrt(p1_H0*(1-p1_H0)*(1/n1 + 1/n2))
se_H1_1 = np.sqrt(p1_H1*(1-p1_H1)/n1)
se_H1_2 = np.sqrt(p2_H1*(1-p2_H1)/n2)

x = np.linspace(0.08, 0.20, 500)


#plot
fig, ax = plt.subplots(1, 2, figsize=(11,4), sharey=True)

# Under H0: same center
y = norm.pdf(x, p1_H0, se_H0)
ax[0].plot(x, y, color="purple", lw=2)
ax[0].fill_between(x, y, color="orchid", alpha=0.4)
ax[0].axvline(p1_H0, color="black", ls="--")
ax[0].set_title("Under H₀ :  p₁ = p₂ = 0.12")
ax[0].set_xlabel("Sample proportion")
ax[0].set_ylabel("Sampling density")
ax[0].grid(alpha=0.2)

# Under H1: two separated distributions
y1 = norm.pdf(x, p1_H1, se_H1_1)
y2 = norm.pdf(x, p2_H1, se_H1_2)
ax[1].plot(x, y1, color="steelblue", lw=2, label="Group A (p₁=0.12)")
ax[1].plot(x, y2, color="tomato", lw=2, label="Group B (p₂=0.16)")
ax[1].fill_between(x, np.minimum(y1,y2), color="gray", alpha=0.5)
ax[1].axvline(p1_H1, color="steelblue", ls="--", alpha=0.7)
ax[1].axvline(p2_H1, color="tomato", ls="--", alpha=0.7)
ax[1].set_title("Under H₁ :  true proportions differ")
ax[1].set_xlabel("Sample proportion")
ax[1].legend(fontsize=8)
ax[1].grid(alpha=0.2)

plt.suptitle("Two-Sample Proportion Test", fontsize=14, y=1.0)
plt.tight_layout()
plt.show()
```

Under the null hypothesis (left plot), both groups have the same true proportion, so their sampling distributions overlap completely and any small difference we observe is just sampling noise.

Under the alternative (right plot), the distributions are separated: the means differ by a real effect (e.g. 0.12 vs 0.16).  
The two-proportion z-test measures *how far apart the observed proportions are* relative to the expected variation under H₀.

→ The greater the separation (measured in standard-error units),
the larger the z-statistic and the smaller the p-value.


<a id="paired-t-test"></a>
## 12. Paired t-Test (Before/After, Matched Pairs)

$\ \ $  
Tests whether the *mean* change within subjects, that measured twice, differes from zero (e.g., before/after) → **dependence** within pairs.

<br>

### Data Structure
We have paired observations $(X_i^{\text{before}}, X_i^{\text{after}})$ on the **same units** (or matched pairs), with paired differences:  
  
$$
D_i = X_i^{\text{after}} - X_i^{\text{before}}, \quad i=1,\dots,n
$$  

Let $\bar D$ be the sample mean of the differences and $s_D$ their sample standard deviation.  

<br>

### Assumptions:
- **Pairs are meaningful** (same subject or well-matched units).  
- **Independence across pairs** (pairs don’t influence each other).  
- **Approx. normal differences** $D_i$ (or $n$ moderately large by CLT).  

<br> 

### Hypotheses
Two-tailed:
$$
H_0:\ \mu_D = 0 \quad\text{vs}\quad H_1:\ \mu_D \neq 0
$$
Right-tailed (improvement/increase): $H_1:\ \mu_D > 0$   
Left-tailed (decrease): $H_1:\ \mu_D < 0$

<br> 

### Test statistic
Under $H_0$, $t \sim t_{n-1}$:  
$$
\begin{align}
t_c = \frac{\bar D}{\,s_D/\sqrt{n}\,}, \qquad \text{df}=n-1,
\quad
\quad 
& \text{if } |t_c| > t_{1-\alpha/2} &\longrightarrow \text{Reject the } H_0 \\
& \text{if } p_{value} < \alpha &\longrightarrow \text{Reject the } H_0
\end{align}
$$


<br> 

### Effect size (paired)
The standardized mean change (Cohen’s $d_z$):
$$
d_z = \frac{\bar D}{s_D}
$$
Conventions: small $\approx 0.2$, medium $\approx 0.5$, large $\approx 0.8$.  


```python
import numpy as np
from scipy import stats

np.random.seed(1)
n = 30
before = np.random.normal(50, 10, size=n)
after  = before - np.random.normal(3, 5, size=n)  # improvement ~ +3

t_stat, p_two = stats.ttest_rel(before, after)

# One-tailed example: after < before (mean difference negative)
diff = after - before
t_stat_one = t_stat
p_left = p_two/2 if t_stat < 0 else 1 - p_two/2

print(f"Paired t: t={t_stat:.3f}, two-tailed p={p_two:.4f}, one-tailed (after<before) p={p_left:.4f}")
print(f"Mean diff = {diff.mean():.2f}")
```

<a id="chi-square-goodness-of-fit"></a>
## 13. Chi-Square Goodness-of-Fit Test

Tests whether the observed counts across $k$ categories follow a **specified theoretical distribution** (e.g. are colors of M&M candies distributed as claimed? / does customer choice follow predicted market shares?). In other words, it measures the deviation between observed & expected counts.
  
### Assumptions
- Observations are independent  
- Categories are mutually exclusive & exhaustive  
- Each expected count $E_i \ge 5$ (for reliable approximation)  
- Total sample size $n$ is fixed

### Hypotheses
$$
H_0:\ \text{observed frequencies match expected probabilities } (p_1,p_2,\ldots,p_k)
$$
$$
H_1:\ \text{at least one category differs from expectation}
$$

### Test statistic
Let $O_i$ = observed count in category $i$, and $E_i=n p_i$ the expected count (where $n=\sum_i O_i$), $k$ the number of categories:

$$
\chi^2_c = \sum_{i=1}^k \frac{(O_i - E_i)^2}{E_i}
$$

Under $H_0$:
$$
\chi^2_c \sim \chi^2_{k-1}
$$
 

If $\chi^2_c >\ \chi^2_{1-\alpha,\,k-1}$  or if $p$-value < $\alpha$ $\longrightarrow$ Reject $H_0$



### Effect size
Use **Cramér’s V** for strength of deviation:
$$
V = \sqrt{\frac{\chi^2}{n\,(k-1)}}
$$  

with conventional interpretation: *small ≈ 0.1, medium ≈ 0.3, large ≈ 0.5*.  
  

### Residuals
Always inspect the **residuals**:
$$
r_i = \frac{O_i - E_i}{\sqrt{E_i}}
$$
to see which categories contribute most to $\chi^2$.

### Visual check (bar chart)
Compare observed vs expected frequencies visually.



```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Example: test whether a six-sided die is fair
observed = np.array([18, 22, 19, 16, 21, 24])  # observed frequencies
expected = np.ones_like(observed) * observed.sum() / len(observed)  # equal probabilities

chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)
n = observed.sum()
k = len(observed)
cramers_v = np.sqrt(chi2_stat / (n * (k - 1)))

print(f"Chi² statistic = {chi2_stat:.3f}")
print(f"df = {k-1}")
print(f"p-value = {p_val:.4f}")
print(f"Cramér's V = {cramers_v:.3f}")

categories = [f"subject {i+1}" for i in range(len(observed))]
x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, observed, width, label="Observed", color="skyblue")
plt.bar(x + width/2, expected, width, label="Expected", color="lightcoral")
plt.title("Chi-Square Goodness-of-Fit: Observed vs Expected")
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.xticks(x, categories)
plt.legend()
plt.show()
```

## 14. Chi-Square Test of Independence

It determines whether two **categorical variables** are statistically **independent** (e.g. gender and movie preference, device and click behavior). It actually measures deviation from independence. 
  
### Conditions
- Observations are independent  
- Categories are mutually exclusive  
- Expected cell counts $E_{ij} \ge 5$ (rule of thumb)  
- Sample size large enough for the $\chi^2$ approximation  
  
### Data structure
An example of the observed or expected results table:

|                | subject 1 | subject 2 | subject 3 | Row Total |
|----------------|-----------|-----------|-----------|-----------|
| Group 1 | a | b | c | a + b + c |
| Group 2 | d | e | f | d + e + f
| **Column Total** | a + d | b + e | c + f | n |

### Hypotheses
$$
H_0:\ \text{the two variables are independent}
$$
$$
H_1:\ \text{the two variables are not independent}
$$

### Test statistic
We start from a **contingency table** of observed frequencies $O_{ij}$  
(rows = categories of variable A, columns = categories of variable B).

Expected counts if variables were independent:
$$
E_{ij} = \frac{(\text{row total})_i \times (\text{column total})_j}{n}
$$

Then compute
$$
\chi^2 = \sum_i \sum_j \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
$$

Under $H_0$:
$$
\chi^2 \sim \chi^2_{(r-1)(c-1)}
$$
where $r =$ number of rows, $c =$ number of columns.

If $\chi^2 > \chi^2_{1-\alpha,\,df}$ or if $p$-value < $\alpha$ $\longrightarrow$ Reject $H_0$  
- A **small** $p$ (e.g. < 0.05) → unlikely to observe such a deviation if variables were independent → reject $H_0$.  
- **Large** $p$ → data consistent with independence.



### Effect size
For larger tables, **Cramér’s V** is a standardized measure of association strength. It quantifies how strong the association is: even with large samples, tiny deviations can become significant but practically negligible:   

$$
V = \sqrt{\frac{\chi^2}{n \times \min(r-1,\;c-1)}}
$$  

with conventional interpretation: *small ≈ 0.1, medium ≈ 0.3, large ≈ 0.5*.  


  
### Residuals
The residuals are a standardized measure of association strength:  

$$
r_{ij} = \frac{O_{ij} - E_{ij}}{\sqrt{E_{ij}}}
$$
Positive $r_{ij}$ → that cell occurs more often than expected.  
Negative $r_{ij}$ → less often than expected.





```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table (e.g. Device × Click)
# Rows: Desktop / Mobile,  Columns: Clicked / Not Clicked
table = np.array([[120, 80],
                  [150,170]])

chi2_stat, p_val, dof, expected = chi2_contingency(table, correction=False)
n = table.sum()
r, c = table.shape
cramers_v = np.sqrt(chi2_stat / (n * min(r-1, c-1)))

print("Observed frequencies:\n", table)
print("\nExpected frequencies under independence:\n", expected.round(2))
print(f"\nChi² = {chi2_stat:.3f},  df = {dof},  p = {p_val:.4f}")
print(f"\Cramér's V = {cramers_v:.3f}  →  {(abs(cramers_v)<0.1 and 'small' or abs(cramers_v)<0.3 and 'medium' or 'large')} effect")

if p_val < 0.05:
    print("Reject H0 → variables are not independent (association exists).")
else:
    print("Do not reject H0 → no evidence of association.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

residuals = (table - expected) / np.sqrt(expected)
labels = ["Desktop", "Mobile"]
cols = ["Clicked", "Not Clicked"]

df_resid = pd.DataFrame(residuals, index=labels, columns=cols)

plt.figure(figsize=(5,4))
sns.heatmap(df_resid, annot=True, cmap="coolwarm", center=0)
plt.title("Standardized Residuals (Observed - Expected)")
plt.show()
```

## 15. Fisher’s Exact Test (2×2 Contingency Tables)

For **small samples** or cells with expected counts below 5, the chi-square test’s approximation to the sampling distribution breaks down.  
  
The **Fisher’s Exact Test** provides an **exact p-value**, without approximation. It is applied when:
- the **sample size is small (any expected cell < 5)** 
- **2×2 categorical data**
- the **exact inference is needed (e.g., clinical trials, rare events)**  
  
***Because the sample is small, Fisher’s test reports exact probabilities, not approximations.***

<br>
<br>

### Data structure
A 2×2 contingency table of observed frequencies:

|               | Success | Failure | Row Total |
|----------------|----------|----------|------------|
| Group 1 | a | b | a + b |
| Group 2 | c | d | c + d |
| **Column Total** | a + c | b + d | n |


<br>
<br>

### Hypotheses
$$
H_0:\ \text{the two groups have the same proportion of successes}
$$
$$
H_1:\ \text{the proportions differ (two-tailed) or one is greater (one-tailed)}
$$

<br>
<br>

### Probability of observing this exact table
Under $H_0$, the marginal totals are fixed.  
The probability of any table with the same marginals follows a **hypergeometric distribution**:  

$$
P(a) = \frac{\dbinom{a+b}{a}\dbinom{c+d}{c}}{\dbinom{n}{a+c}}
$$  


The p-value is obtained by summing the probabilities of all tables that are as or more extreme than the observed one, in the direction of $H_1$.  
  
- **p-value < α** → reject $H_0$: the proportions differ significantly between groups.  
- **p-value ≥ α** → insufficient evidence of difference.
  
<br>
<br>

The **odds ratio (OR)** quantifies the strength of association:
$$
\text{OR} = \frac{a\,d}{b\,c}
$$  

An OR > 1 indicates higher odds of success in Group 1;
OR < 1 indicates lower odds.

<br>
<br>

**Rule of thumb**  

| Situation | Preferred test |
|------------|----------------|
| Any expected cell < 5 | Fisher’s Exact Test |
| All expected ≥ 5 | Chi-Square Test |
| Large sample | Chi-Square (faster) |
| 2×2 small sample | Fisher (exact) |



```python
import numpy as np
from scipy.stats import fisher_exact

# 2x2 contingency table
# Example: Treatment × Outcome
# Rows: Treatment / Control,  Columns: Success / Failure
table = np.array([[5, 1],
                  [2, 6]])

# Two-sided Fisher’s Exact Test
odds_ratio, p_two = fisher_exact(table, alternative='two-sided')

# One-tailed examples
p_right = fisher_exact(table, alternative='greater')[1]  # more successes in first row
p_left  = fisher_exact(table, alternative='less')[1]     # fewer successes in first row

print("Observed table:\n", table)
print(f"\nOdds ratio = {odds_ratio:.3f}")
print(f"Two-tailed p = {p_two:.4f}")
print(f"Right-tailed p = {p_right:.4f}   |   Left-tailed p = {p_left:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
plt.axhline(1, color="gray", ls="--")
plt.scatter(1, odds_ratio, color="tomato", s=80)
plt.text(1.05, odds_ratio, f"OR = {odds_ratio:.2f}", va="center")
plt.title("Fisher's Exact Test — Odds Ratio")
plt.ylabel("Observed Odds Ratio")
plt.xlabel("Reference (OR = 1 under H₀)")
plt.ylim(0, max(2, odds_ratio+0.5))
plt.show()
```

# V. ANOVA  

ANOVA stands for Analysis of Variance, because instead of comparing means directly, we analyze how variance is split into different sources:

$$
\text{Total variance = Between-group variance + Within-group variance}
$$  
  
  
So ANOVA tests whether the *means* of **three or more independent groups** are all equal  
and answers to the question: ***“How much of the total variability in my data can be explained by the group factor?”***  
  
*When there are only two groups, ANOVA and the two-sample $t$-test give identical results.*  

<br>

- **One-Way ANOVA:** only one factor influences the outcome. We test if means differ between groups.

- **Two-Way ANOVA:** two factors influence the outcome, and we can test:
  1. The main effect of each factor (Factor1, Factor2)
  2. Their interaction effect (Factor1 × Factor2)

So “one-way” = one explanatory variable (one factor) used to form the groups being compared,  
while “two-way” = two explanatory variables, possibly interacting.

<br>

<a id="one-way-anova"></a>
## 16. One-Way ANOVA

In an one-way ANOVA, there is **only one independent variable (factor) that defines the groups**.
  
>e.g. Let’s say we study the effect of teaching method (A, B or C) on exam scores. Here, the factor “Method” has 3 levels: A, B, C. We’re testing whether the mean score differs between these 3 groups.  
There’s only one factor (Method) that created the groups → One-Way ANOVA.  


### Assumptions
- Groups are independent  
- Each group is normally distributed  
- Homogeneity of variances (equal $\sigma^2$ across groups)

*If the third assumption fails, we have to use **Welch’s ANOVA** instead.*

<br>

### Hypotheses
$$
H_0:\ \mu_1 = \mu_2 = \dots = \mu_k 
\quad
\text{vs}
\quad
H_1:\ \text{at least one mean differs}
$$

<br>

### Logic
If all group means are equal, then the **variation between groups** should be comparable to the **variation within groups** (due to random noise).

We split the total variability (sum of squares) into:
$$
SS_T = SS_B + SS_W
$$

- $SS_B$: Between-groups sum of squares  
  $$
  SS_B = \sum_{i=1}^k n_i (\bar X_i - \bar X)^2
  $$
- $SS_W$: Within-groups sum of squares  
  $$
  SS_W = \sum_{i=1}^k \sum_{j=1}^{n_i} (X_{ij} - \bar X_i)^2
  $$

<br>

### Test statistic
The **F-statistic** compares these two sources of variation:
$$
F = \frac{MS_B}{MS_W} = 
\frac{SS_B/(k-1)}{SS_W/(N-k)}
$$
Under $H_0$:  

$$
F \sim F_{k-1,\ N-k}, \quad \quad \text{if } F > F_{1-\alpha,\ k-1,\ N-k} \longrightarrow \text{Regect } H_0
$$  
  
- A large $F$ statistic (and small $p$) → unlikely that all group means are equal → reject $H_0$.
- A small $F$ (large $p$) → differences among sample means are consistent with random variation.

<br>

### Effect size
The **eta-squared** measure indicates the proportion of total variance explained by group differences:  

$$
\eta^2 = \frac{SS_B}{SS_T}
$$

with conventional interpretation: *small ≈ 0.01, medium ≈ 0.06, large ≈ 0.14*.  

<br>

### Post-hoc tests
If the overall ANOVA is significant, we need to identify *which* groups differ.
Common multiple-comparison procedures:
- **Tukey’s HSD** (controls family-wise error)
- **Bonferroni** correction
- **Holm’s** method

These compare all pairs of means while adjusting $p$-values.



```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

np.random.seed(42)

# Simulated data: three treatments with different means
g1 = np.random.normal(50, 5, 30)
g2 = np.random.normal(52, 5, 30)
g3 = np.random.normal(56, 5, 30)

df = pd.DataFrame({
    "value": np.r_[g1, g2, g3],
    "group": ["A"]*len(g1) + ["B"]*len(g2) + ["C"]*len(g3)
})

# One-way ANOVA
model = ols("value ~ C(group)", data=df).fit()
anova_table = anova_lm(model, typ=2)
print(anova_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="group", y="value", palette="pastel")
plt.title("One-Way ANOVA — Group Distributions")
plt.xlabel("Group")
plt.ylabel("Measured value")
plt.show()
```

## 17. Two-Way ANOVA

The **Two-Way ANOVA** examines the influence of **two independent factors** on one continuous dependent variable.

It also allows testing whether the factors **interact** that is, whether the effect of one factor depends on the level of the other.

>e.g. Suppose we measure exam scores based on **Factor A:** Teaching Method (A₁, A₂, A₃) and **Factor B:** Gender (Male, Female)
>
>We can test:
>1. The main effect of Method  
>2. The main effect of Gender  
>3. The **interaction effect** (Method × Gender)

### Assumptions
- Independence of observations  
- Normality within each cell  
- Equal variances across all factor combinations


### Model
Let $Y_{ijk}$ denote the observation for level $i$ of factor A,
level $j$ of factor B, and replication $k$.

The full factorial model is:

$$
Y_{ijk} = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \varepsilon_{ijk}
$$

where:
- $\mu$ = overall mean  
- $\alpha_i$ = effect of level $i$ of Factor A  
- $\beta_j$ = effect of level $j$ of Factor B  
- $(\alpha\beta)_{ij}$ = interaction effect between A and B  
- $\varepsilon_{ijk} \sim N(0, \sigma^2)$ = random error

<br>

### Hypotheses

1. **Main effect of A (Factor A):**
   $$
   H_0^A:\ \alpha_1 = \alpha_2 = \dots = 0
   $$

2. **Main effect of B (Factor B):**
   $$
   H_0^B:\ \beta_1 = \beta_2 = \dots = 0
   $$

3. **Interaction (A×B):**
   $$
   H_0^{A\times B}:\ (\alpha\beta)_{ij} = 0\ \forall\ i,j
   $$

Rejecting the interaction null means
“the effect of one factor depends on the level of the other.”

<br>

### Test Statistics
The ANOVA output table includes three $F$-tests:

| Source | Meaning |
|---------|----------|
| C(Factor_A) | Main effect of Factor A |
| C(Factor_B) | Main effect of Factor B |
| C(Factor_A) : C(Factor_B) | Interaction effect between A and B |

- A **significant main effect** means that factor changes the mean of $Y$ overall.  
- A **significant interaction** means that the effect of one factor depends on the level of the other → interpret the **interaction first** before the main effects.


If the interaction term has **$p < 0.05$** , then **the effect of Factor A on the output depends on Factor B**

If interaction is not significant **$p > 0.05$** , then we can **interpret the main effects separately**.  

<br>

### Effect size for Two-Way ANOVA
Use **partial eta-squared** ($\eta_p^2$) for each factor is interpreted as the proportion of variance in $Y$ explained by that factor alone:  

$$
\eta_p^2 = \frac{SS_{\text{effect}}}{SS_{\text{effect}} + SS_{\text{error}}}
$$  


with conventional interpretation: *small ≈ 0.01, medium ≈ 0.06, large ≈ 0.14*.  

```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

np.random.seed(123)

# Factors: Factor A (A1, A2, A3) × Factor B (B1, B2)
A = np.repeat(["A1", "A2", "A3"], 40)
B = np.tile(np.repeat(["B1", "B2"], 20), 3)

# Define a model with main effects and interaction
mu = 70
effects_A = {"A1": 0, "A2": 5, "A3": 0}
effects_B = {"B1": 0, "B2": 3}

# Interaction: A2 × B2 adds +4 bonus
interaction_effect = np.array([(a == "A2" and b == "B2") * 4 for a, b in zip(A, B)])

# Response variable with noise
Y = mu + np.array([effects_A[a] for a in A]) + np.array([effects_B[b] for b in B]) + interaction_effect + np.random.normal(0, 5, len(A))

# Build DataFrame
df = pd.DataFrame({"Mean": Y, "Factor_A": A, "Factor_B": B})

# Two-way ANOVA with interaction
model = ols("Mean ~ C(Factor_A) * C(Factor_B)", data=df).fit()
anova_tbl = anova_lm(model, typ=2)
print(anova_tbl)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.pointplot(data=df, x="Factor_A", y="Mean", hue="Factor_B", dodge=True, markers=["o","s"], capsize=.1)
plt.title("Two-Way ANOVA — Interaction Plot")
plt.xlabel("Factor A")
plt.ylabel("Mean")
plt.show()
```

<a id="normality-tests"></a>
## 18. Normality Tests

Many parametric tests (e.g., $t$-tests, ANOVA) assume that data are **approximately normal** (within each group). Normality tests assess whether a sample plausibly comes from a normal distribution.

**Important:** With large $n$, tiny deviations can become “significant.” Always combine formal tests with **graphics** (histogram, QQ plot) and **practical judgment**.



### Main tests

1. **Shapiro–Wilk**  
   Very powerful for small–moderate $n$ (default choice in many workflows).
   - $H_0$: data are i.i.d. normal  
   - Statistic: $W$ (close to 1 for normal data)  
   - Small $p \Rightarrow$ reject $H_0$  
   
   
2. **D’Agostino–Pearson $K^2$**  
   Uses skewness and kurtosis jointly (good all-round; needs moderate $n$).
   - $H_0$: data are normal  
   - Statistic: $K^2$ (approximately $\chi^2_2$ under $H_0$)  
   
   
3. **Kolmogorov–Smirnov (KS)**  
   Compares empirical CDF vs theoretical CDF.  
   **Caveat:** If you **estimate** $\mu,\sigma$ from the same data, the vanilla KS test is **not exact** (this is the *Lilliefors* situation). Prefer other tests or use an adjusted Lilliefors test (not in SciPy core).  
   
   
4. **Anderson–Darling (A–D)**  
   Gives **critical values** for normality and emphasizes tails more than KS.  
   No $p$-value; compare the statistic with tabulated critical values.

<br>

### Typical guidance
- Small–moderate samples ($n \le 500$): prefer **Shapiro–Wilk**.  
- Larger samples: **D’Agostino $K^2$** works well.  
- Always add a **QQ plot** and **histogram**.

<br>

### Decision logic
- If **normality holds** (and variances are roughly equal), parametric tests are appropriate.
- If **normality is doubtful** (heavy skew/outliers), consider transformations (e.g., log, sqrt) or **nonparametric** tests (Wilcoxon, Mann–Whitney, Kruskal–Wallis).



```python
# Single vector Normality diagnostics tool

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def normality_report(x, alpha=0.05, title="Sample"):
    x = np.asarray(x)
    x = x[~np.isnan(x)]
    n = len(x)
    print(f"{title}: n={n}")

    # Shapiro–Wilk
    W, p_sw = stats.shapiro(x)
    print(f"Shapiro–Wilk: W={W:.3f}, p={p_sw:.4f}  -> {'Reject H0' if p_sw<alpha else 'Do not reject H0'}")

    # D’Agostino–Pearson K^2
    K2, p_k2 = stats.normaltest(x)  # combines skewness & kurtosis
    print(f"D’Agostino–Pearson K^2: K2={K2:.3f}, p={p_k2:.4f}  -> {'Reject H0' if p_k2<alpha else 'Do not reject H0'}")

    # KS against fitted Normal (note Lilliefors caveat)
    mu, sigma = x.mean(), x.std(ddof=1)
    D, p_ks = stats.kstest((x - mu)/sigma, 'norm')
    print(f"Kolmogorov–Smirnov (vs fitted Normal): D={D:.3f}, p={p_ks:.4f}  [not exact if params estimated]")

    # Anderson–Darling
    ad = stats.anderson(x, dist='norm')
    print(f"Anderson–Darling: A2={ad.statistic:.3f}")
    for sig, crit in zip(ad.significance_level, ad.critical_values):
        verdict = "Reject H0" if ad.statistic > crit else "Do not reject H0"
        print(f"  at {sig:.0f}%: critical={crit:.3f} -> {verdict}")

    # histogram
    plt.figure(figsize=(6,4))
    plt.hist(x, bins=max(10, int(np.sqrt(n))))
    plt.title(f"Histogram - {title}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

    # QQ plot
    plt.figure(figsize=(6,4))
    stats.probplot(x, dist="norm", plot=plt)
    plt.title(f"QQ plot - {title}")
    plt.tight_layout()
    plt.show()

    
# Example usage:
x = np.random.lognormal(mean=0.0, sigma=0.6, size=120)
normality_report(x, title="Example data")
```

```python
# Groupwise Normality diagnostics tool

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math


def normality_by_group(df, value_col, group_col, alpha=0.05):
    """
    Performs Shapiro–Wilk and D’Agostino–Pearson K^2 normality tests for each group in the DataFrame.

    Parameters
    ----------
    df : DataFrame
         Your dataset containing the numeric column and group labels.
    value_col : str
         Name of the numeric column to test.
    group_col : str
         Name of the grouping (categorical) column.
    alpha : float, optional
         Significance level for flagging rejections (default=0.05).
    """
    
    out = []
    for g, sub in df.groupby(group_col):
        x = sub[value_col].dropna().to_numpy()
        n = len(x)
        if n < 3:
            out.append((g, n, np.nan, np.nan, np.nan, np.nan))
            continue
            
        # Shapiro–Wilk (good for small–moderate n)
        W, p_sw = stats.shapiro(x)
        
        # D’Agostino–Pearson K^2 (needs n >= 8)
        if n >= 8:
            K2, p_k2 = stats.normaltest(x)
        else:
            K2, p_k2 = np.nan, np.nan
        out.append((g, n, W, p_sw, K2, p_k2))
        
    res = pd.DataFrame(out, columns=[group_col, "n", "ShapiroWilk", "ShapiroWilk_p", "K2", "K2_p"])
    res["Normal_Shapiro"] = np.where(res["ShapiroWilk_p"] < alpha, "non normal", "normal")
    res["Normal_K2"] = np.where(res["K2_p"] < alpha, "non normal", "normal")
    
    return res



def boxplots_by_group(df, value_col, group_col, order=None, title="Boxplots by group"):
    
    groups = sorted(df[group_col].unique()) if order is None else order
    data = [df.loc[df[group_col] == g, value_col].dropna().values for g in groups]
    
    plt.figure(figsize=(6 + 0.4*len(groups), 4))
    plt.boxplot(data, labels=groups, showmeans=True)
    plt.title(title)
    plt.xlabel(group_col)
    plt.ylabel(value_col)
    plt.tight_layout()
    plt.show()



def qqplots_by_group(df, value_col, group_col, order=None, suptitle="QQ plots by group"):
    
    groups = sorted(df[group_col].unique()) if order is None else order
    k = len(groups)
    ncols = int(math.ceil(math.sqrt(k)))
    nrows = int(math.ceil(k / ncols))
    
    fig, axes = plt.subplots(nrows, ncols, figsize=(4*ncols, 3.5*nrows))
    axes = np.atleast_1d(axes).ravel()
    for ax, g in zip(axes, groups):
        x = df.loc[df[group_col] == g, value_col].dropna().values
        stats.probplot(x, dist="norm", plot=ax)
        ax.set_title(f"{group_col} = {g}")
        # make points hollow for readability
        try:
            ax.get_lines()[0].set_markerfacecolor("none")
        except Exception:
            pass
    for ax in axes[len(groups):]:
        ax.axis("off")
    fig.suptitle(suptitle, fontsize=12)
    plt.tight_layout()
    plt.show()

    
    
def group_diagnostics(df, value_col, group_col, alpha=0.05, order=None, do_box=True, do_qq=True, title_prefix=None):
    """
    Run groupwise normality tests + optional boxplots and QQ plots.

    Returns
    -------
    pandas.DataFrame
        Table with Shapiro–Wilk and D’Agostino K^2 per group (+ flags).
    """
    
    title_prefix = "" if title_prefix is None else f"{title_prefix} -  "
    res = normality_by_group(df, value_col, group_col, alpha=alpha)
    if do_box:
        boxplots_by_group(df, value_col, group_col, order=order,
                          title=f"{title_prefix}Boxplots by {group_col}")
    if do_qq:
        qqplots_by_group(df, value_col, group_col, order=order,
                         suptitle=f"{title_prefix}QQ plots by {group_col}")
    return res



# Example
np.random.seed(123)
df_example = pd.DataFrame({
     "value": np.r_[np.random.normal(50, 5, 30),
                    np.random.normal(52, 5, 30),
                    np.random.lognormal(3.8, 0.3, 30)],
     "group": ["A"]*30 + ["B"]*30 + ["C"]*30
 })


res = group_diagnostics(df_example, "value", "group", alpha=0.05, title_prefix="Diagnostics")
res
```

## 19. Homogeneity of Variance Tests (Levene, Brown–Forsythe, Bartlett)

Many parametric tests (two-sample $t$, one-way ANOVA) assume **equal population variances** across groups:  

$$
H_0:\ \sigma_1^2 = \sigma_2^2 = \dots = \sigma_k^2
$$  

If this assumption fails (heteroscedasticity), we should switch to **Welch’s** variants (Welch’s $t$ or Welch’s ANOVA) or use nonparametric tests.



### Diagnostic Tests

1. **Levene’s test** (robust, uses deviations from a chosen center)
   - Default version uses the **mean**:
     $$
     W = \frac{(N-k)}{(k-1)}\cdot\frac{\sum_{i=1}^k n_i(\bar Z_{i\cdot}-\bar Z_{\cdot\cdot})^2}{\sum_{i=1}^k\sum_{j=1}^{n_i}(Z_{ij}-\bar Z_{i\cdot})^2},
     \quad Z_{ij} = |X_{ij} - \text{center}_i|
     $$  
     
   - **Brown–Forsythe** variant (more robust to non-normality) uses the **median** as center.

   - Hypotheses:
     $$
     H_0:\ \sigma_1^2=\dots=\sigma_k^2 \quad\text{vs}\quad H_1:\ \text{not all equal}
     $$
   - Large $W$ (small $p$) $\Rightarrow$ reject homogeneity.  
   

2. **Bartlett’s test** (most powerful under normality; sensitive to non-normality)
   - Hypotheses same as above.
   - Prefer **only** when data are plausibly normal in all groups.

<br>

### Practical guidance
- If normality is **uncertain** or you have outliers $\Rightarrow$ use **Levene (median)** a.k.a. **Brown–Forsythe**.
- If normality looks **good** (QQ plots, Shapiro) $\Rightarrow$ Bartlett is fine and slightly more powerful.
- If $p < \alpha$: $\ \longrightarrow$ variances are unequal (heteroscedasticity):
  - Two groups: use **Welch’s $t$-test** (set `equal_var=False`).
  - $k\ge 3$ groups: use **Welch’s ANOVA**.




```python
import numpy as np
from scipy import stats

# Example groups (replace with your arrays)
np.random.seed(7)
g1 = np.random.normal(50, 5, 30)
g2 = np.random.normal(52, 8, 28)   # larger variance
g3 = np.random.normal(56, 5, 32)

# Levene's test (center='mean' is classical Levene)
W_mean, p_mean = stats.levene(g1, g2, g3, center='mean')

# Brown–Forsythe (Levene with median center)
W_med, p_med = stats.levene(g1, g2, g3, center='median')

# Bartlett (powerful if normal, sensitive otherwise)
B, p_bart = stats.bartlett(g1, g2, g3)

print(f"Levene (mean):    W={W_mean:.3f}, p={p_mean:.4f}")
print(f"Brown–Forsythe:   W={W_med:.3f},  p={p_med:.4f}")
print(f"Bartlett:         B={B:.3f},  p={p_bart:.4f}")
```

```python
from scipy import stats
import pandas as pd
from statsmodels.stats.oneway import anova_oneway  # Welch & Brown–Forsythe ANOVA

# --- Welch's two-sample t-test (independent samples) ---
t_welch, p_welch = stats.ttest_ind(g1, g2, equal_var=False)
print(f"Welch two-sample t-test: t={t_welch:.3f}, p={p_welch:.4f}")

# --- Welch's one-way ANOVA for k groups ---
# statsmodels' anova_oneway provides unequal-variance ANOVA (Welch) and Brown–Forsythe
welch_res = anova_oneway([g1, g2, g3], use_var='unequal')  # Welch ANOVA
bf_res    = anova_oneway([g1, g2, g3], use_var='bf')       # Brown–Forsythe ANOVA

print("\nWelch ANOVA:")
print(welch_res)  # prints statistic, p-value, df, etc.

print("\nBrown–Forsythe ANOVA:")
print(bf_res)
```

```python
def variance_diagnostics(*groups, alpha=0.05, assume_normal=False):
    """
    Runs Levene (median), Levene (mean), and Bartlett.
    Returns a dict with a recommendation for follow-up test.
    """
    # Convert to list of 1D arrays without NaNs
    G = [np.asarray(g)[~np.isnan(g)] for g in groups if len(g)]
    k = len(G)
    n_total = sum(len(g) for g in G)

    # Tests
    W_med, p_med = stats.levene(*G, center='median')
    W_mean, p_mean = stats.levene(*G, center='mean')
    B, p_bart = stats.bartlett(*G)

    # Decision
    equal_var = (p_med >= alpha) and (p_mean >= alpha) and (assume_normal and p_bart >= alpha or True)

    if equal_var:
        rec = "Equal variances plausible → use standard tests (Student t / one-way ANOVA)."
    else:
        if k == 2:
            rec = "Heteroscedastic → use Welch's two-sample t-test (equal_var=False)."
        else:
            rec = "Heteroscedastic → use Welch's ANOVA (use_var='unequal') or Brown–Forsythe (use_var='bf')."

    return {
        "k_groups": k, "n_total": n_total,
        "Levene_median_W": W_med, "Levene_median_p": p_med,
        "Levene_mean_W": W_mean, "Levene_mean_p": p_mean,
        "Bartlett_B": B, "Bartlett_p": p_bart,
        "recommendation": rec
    }

diag = variance_diagnostics(g1, g2, g3, alpha=0.05, assume_normal=False)
for k, v in diag.items():
    if isinstance(v, float):
        print(f"{k}: {v:.4f}")
    else:
        print(f"{k}: {v}")
```

- **Robust default:** Prefer **Levene (median)** / Brown–Forsythe when normality is doubtful.
- **Pipeline order:** Run **normality checks (Section 19)** → then **variance homogeneity** (this section) → pick the correct test (standard vs Welch).
- **Report clearly:** State which variance test you used and how it informed your choice of $t$/ANOVA variant.
- **Multiple groups with very unequal $n$:** Welch’s ANOVA is especially helpful (less sensitive to imbalance + heteroscedasticity).

## 20. Correlation Tests (Pearson, Spearman, Kendall)

Correlation tests measure the **strength and direction** of the relationship between two numeric variables.


### Pearson’s correlation
- Measures **linear association** between $X$ and $Y$.
- Assumes both variables are **continuous**, **normally distributed**, and the relationship is **linear**.
- Statistic:
  $$
  r = \frac{\sum (x_i - \bar x)(y_i - \bar y)}{\sqrt{\sum (x_i - \bar x)^2}\sqrt{\sum (y_i - \bar y)^2}}
  $$
- Range: $r \in [-1, 1]$

  - $r > 0$: positive correlation  
  - $r < 0$: negative correlation  
  - $|r| \approx 1$: strong linear relationship  
  - $|r| \approx 0$: weak or no linear relationship


### Spearman’s rank correlation
- Nonparametric alternative to Pearson.  
- Uses **ranks** instead of raw values.
- Detects **monotonic** relationships (not necessarily linear).
- Statistic:
  $$
  \rho = 1 - \frac{6\sum d_i^2}{n(n^2 - 1)}
  $$
  where $d_i$ = difference between ranks of $x_i$ and $y_i$.


### Kendall’s tau
- Also rank-based; measures **ordinal association**.
- More robust with small $n$ or many ties.
- Statistic:
  $$
  \tau = \frac{n_c - n_d}{\tfrac{1}{2}n(n-1)}
  $$
  where $n_c$ = number of concordant pairs, $n_d$ = discordant pairs.


### Hypotheses
For all correlation tests:

$$
H_0: \text{variables are independent (no association)}
\quad
\text{vs}
\quad
H_1$: \text{variables are associated } (\rho \neq 0)
$$



### Choice summary

| Test | Type | Assumptions | Detects |
|------|------|--------------|----------|
| Pearson | Parametric | Linear, normal | Linear correlation |
| Spearman | Nonparametric | Monotonic | Rank correlation |
| Kendall | Nonparametric | Ordinal, ties OK | Concordance |


### Interpretation guidelines

| | Very weak | Weak | Moderate | Strong | Very strong |
|--|-----------|-------|-----------|----------|--------------|
| **|r|** | < 0.2 | 0.2 – 0.4 | 0.4 – 0.6 | 0.6 – 0.8 | > 0.8 |


### Visualization
Always inspect the scatterplot:
- Curved trend → use **Spearman**
- Linear pattern → **Pearson**
- If distributions are skewed or contain outliers, report both Pearson and Spearman for robustness
- Outliers → check influence (they can flip sign).

<br>

***Notes***
- Correlation ≠ causation.
- For categorical × categorical association → use **Chi-square** or **Cramér’s V**.
- For partial correlation (controlling other variables) → use `pingouin.partial_corr()` or `statsmodels`.



```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Example data
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = 2*x + np.random.normal(0, 3, 50)  # linear with noise
y2 = np.log1p(x) + np.random.normal(0, 0.1, 50)  # monotonic but nonlinear

# Pearson correlation (linear)
r, p = stats.pearsonr(x, y)
print(f"Pearson r = {r:.3f}, p = {p:.4f}")

# Spearman rank (monotonic)
rho, p_s = stats.spearmanr(x, y2)
print(f"Spearman ρ = {rho:.3f}, p = {p_s:.4f}")

# Kendall tau (ordinal association)
tau, p_t = stats.kendalltau(x, y2)
print(f"Kendall τ = {tau:.3f}, p = {p_t:.4f}")

# Visualize
fig, ax = plt.subplots(1,2, figsize=(9,4))
ax[0].scatter(x, y)
ax[0].set_title("Linear relationship (Pearson)")
ax[0].set_xlabel("X"); ax[0].set_ylabel("Y")

ax[1].scatter(x, y2, color='orange')
ax[1].set_title("Monotonic, non-linear (Spearman)")
ax[1].set_xlabel("X"); ax[1].set_ylabel("Y₂")
plt.tight_layout()
plt.show()
```

## 21. Power and Sample Size

Statistical **power** is the probability that a test correctly rejects $H_0$ when a real effect exists.  
It answers: “If there *is* a true difference, how likely am I to detect it?”


### Key concepts

- **Significance level ($\alpha$):** probability of *Type I error* (false positive).  
- **Power ($1-\beta$):** probability of *detecting* a true effect (1 – Type II error).  
- **Effect size ($d, r, \eta^2$ etc.):** standardized magnitude of the true effect.  
- **Sample size ($n$):** number of observations per group.


### Relationship
For a given test type:
$$
\text{Power} = f(\text{Effect Size},\, n,\, \alpha,\, \text{Test Type})
$$

Increasing $n$ or $\alpha$, or having a larger effect size, increases power.



### Rules of thumb

| Desired power | Typical choice of α | Required effect size for small n |
|:--:|:--:|:--:|
| 0.8 (80%) | 0.05 | Needs moderate/large effect |
| 0.9 (90%) | 0.05 | Needs smaller effect or larger

### Power analysis in other contexts

| Test | Power function |
|------|----------------|
| One-sample / paired t | `TTestPower()` |
| Two-sample t | `TTestIndPower()` |
| Proportion tests | `NormalIndPower()` |
| ANOVA (one-way) | `FTestAnovaPower()` |
| Correlation | `NormalIndPower()` or simulation |


```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
from statsmodels.stats.power import TTestIndPower

# Initialize power analysis object
analysis = TTestIndPower()

# Parameters
alpha = 0.05
target_power = 0.8
effects = [0.2, 0.5, 0.8]     # Cohen's d: small, medium, large
labels  = ["Small (d=0.2)", "Medium (d=0.5)", "Large (d=0.8)"]
colors  = ["tab:blue", "tab:orange", "tab:green"]

# Compute required sample size per group for each effect
required_ns = []
for d in effects:
    n_req = analysis.solve_power(effect_size=d, power=target_power, alpha=alpha)
    required_ns.append(ceil(n_req))

    
# Range of n for the power curve
n = np.arange(5, 450, 1)

# Plot
plt.figure(figsize=(8,5))
for d, lbl, c, n_req in zip(effects, labels, colors, required_ns):
    power_vals = analysis.power(effect_size=d, nobs1=n, alpha=alpha)
    plt.plot(n, power_vals, color=c, lw=2, label=f"{lbl}")
    
    # vertical line and marker at required n
    plt.axvline(n_req, color=c, ls=":", lw=1)
    p_at_n = analysis.power(effect_size=d, nobs1=n_req, alpha=alpha)
    plt.scatter([n_req], [p_at_n], color=c, edgecolor='black', zorder=3)
    plt.text(n_req+4, p_at_n+0.03, f"n ≈ {n_req}", color=c, fontsize=9)

# target power line
plt.axhline(target_power, color='gray', ls='--', lw=1)
plt.text(100, target_power-0.05, "80% target", color='gray')

# axes and title
plt.ylim(0, 1.05)
plt.xlim(0, 450)
plt.xlabel("Sample size per group", fontsize=11)
plt.ylabel("Statistical Power", fontsize=11)
plt.title("Power vs. Sample Size for Different Effect Sizes\n(two-sample t-test, α = 0.05)", fontsize=13)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# Summary
print(f"Required sample size (n per group) for 80% power at α = {alpha}:")
for lbl, n_req in zip(labels, required_ns):
    print(f"  Effect size {lbl} → required n per group ≈ {n_req}")
```

### Interpretation
- For a medium effect ($d=0.5$) and α = 0.05, you need ≈ 64 observations per group for 80 % power.
- Doubling sample size increases power quickly for small effects but less so for large effects.
- Always report effect size and power together, not only $p$-values.


#### Quick planning example

> "We expect a medium effect ($d = 0.5$).  
> With α = 0.05 and power = 0.8, we’ll need ≈ 64 per group.  
> We’ll plan for 70 to allow for dropouts."


```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
from statsmodels.stats.power import TTestIndPower

# Initialize analysis
analysis = TTestIndPower()

# Parameters
alpha = 0.05
target_power = 0.80
effect_grid = np.linspace(0.05, 1.2, 300)
n_list = [30, 64, 100, 150]   # per group
colors = ["tab:blue", "tab:orange", "tab:green", "tab:red"]

# Required effect sizes
results = []
for n in n_list:
    d_req = analysis.solve_power(effect_size=None, nobs1=n, alpha=alpha, power=target_power)
    results.append((n, d_req))

# Plot
plt.figure(figsize=(8,5))

for (n, d_req), color in zip(results, colors):
    power_curve = analysis.power(effect_size=effect_grid, nobs1=n, alpha=alpha)
    plt.plot(effect_grid, power_curve, color=color, lw=2, label=f"n per group = {n}")

    # vertical line, dot, label
    plt.axvline(d_req, color=color, ls=":", lw=1)
    p_at_d = analysis.power(effect_size=d_req, nobs1=n, alpha=alpha)
    plt.scatter([d_req], [p_at_d], color=color, edgecolor='black', zorder=3)
    plt.text(d_req+0.02, min(0.98, p_at_d+0.04),
             f"d≈{d_req:.2f}", color=color, ha="left", va="bottom", fontsize=8)

# Gray horizontal 80% line
plt.axhline(target_power, color='gray', ls='--', lw=1)
plt.text(effect_grid[0]+0.02, target_power+0.02, "80% target", color='gray', fontsize=9)

# Axes and layout
plt.xlim(effect_grid.min(), effect_grid.max())
plt.ylim(0, 1.1)
plt.xlabel("Effect size (Cohen's d)", fontsize=11)
plt.ylabel("Statistical Power", fontsize=11)
plt.title("Power vs. Effect Size for Different Sample Sizes\n(two-sample t-test, α = 0.05)", fontsize=13)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Summary
print("Required effect size (Cohen's d) for 80% power at α = 0.05:")
for n in n_list:
    d_req = analysis.solve_power(effect_size=None, nobs1=n, alpha=alpha, power=target_power)
    print(f"  n per group = {n:>3}  →  d ≈ {d_req:.3f}")
```

## 22. Effect Sizes


While $p$-values answer “**is there evidence of an effect?**”, effect sizes answer “**how big is it?**”.
They are **scale-free** (comparable across studies) and should be reported **with** confidence intervals when possible.


### 22.1 Cohen’s d - standardized mean difference
For **two independent groups**:
$$
d = \frac{\bar X_1 - \bar X_2}{s_p}, \quad
s_p = \sqrt{\frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1+n_2-2}}
$$

For **paired samples** (within-subjects), use differences $D_i$:
$$
d_z = \frac{\bar D}{s_D}
$$

**Rules of thumb** (Cohen):
small $\approx 0.2$, medium $\approx 0.5$, large $\approx 0.8$.

<br>

### 22.2 Correlation effect size $r$
For **Pearson correlation**, the effect size is simply $|r|$.

From a **$t$-test** (two groups) you can convert to $r$:
$$
r = \sqrt{\frac{t^2}{t^2 + \text{df}}}
$$


**Rules of thumb** (Cohen):
small $\approx 0.10$, medium $\approx 0.30$, large $\approx 0.50$.

<br>

### 22.3 ANOVA effect sizes - $\eta^2$ and partial $\eta_p^2$
For **one-way ANOVA**:
$$
\eta^2 = \frac{SS_{\text{between}}}{SS_{\text{total}}}
$$
For **multi-factor ANOVA**, use **partial eta-squared** for each effect:
$$
\eta_p^2 = \frac{SS_{\text{effect}}}{SS_{\text{effect}} + SS_{\text{error}}}
$$

**Rules of thumb**:
small $\approx 0.01$, medium $\approx 0.06$, large $\approx 0.14$.

<br>

### 22.4 Cramér’s V - strength of association in contingency tables
For **$\chi^2$ tests** (independence) with $r \times c$ table:
$$
V = \sqrt{\frac{\chi^2}{n \cdot \min(r-1,\,c-1)}}
$$

**Rules of thumb** (for df = 1 baseline):
small $\approx 0.10$, medium $\approx 0.30$, large $\approx 0.50$.

<br>

### 22.5 Cohen’s h - difference between two proportions
Used in **two-proportion tests**:
$$
h = 2\Big(\arcsin\sqrt{p_1} - \arcsin\sqrt{p_2}\Big)
$$
**Rules of thumb**: small $\approx 0.20$, medium $\approx 0.50$, large $\approx 0.80$.

<br>

***Reporting tip***: *Always pair **statistical significance** with **effect size** and, when possible, a **confidence interval** for the effect size.*


```python
# Cohen’s d (independent & paired)

import numpy as np
from scipy import stats

# Example independent groups
np.random.seed(0)
x = np.random.normal(50, 10, 80)
y = np.random.normal(55, 10, 75)

# pooled SD
def cohens_d_independent(x, y):
    x, y = np.asarray(x), np.asarray(y)
    nx, ny = len(x), len(y)
    vx, vy = x.var(ddof=1), y.var(ddof=1)
    sp = np.sqrt(((nx-1)*vx + (ny-1)*vy) / (nx+ny-2))
    d = (x.mean() - y.mean()) / sp
    return d

d_ind = cohens_d_independent(x, y)
t_stat, p_val = stats.ttest_ind(x, y, equal_var=True)
print(f"Cohen's d (independent) = {d_ind:.3f}")
print(f"t = {t_stat:.3f}, p = {p_val:.4f}")

# Example paired groups
a = np.random.normal(60, 8, 30)
b = a + np.random.normal(-3, 5, 30)  # after < before
diff = b - a
d_paired = diff.mean() / diff.std(ddof=1)  # d_z
t_paired, p_paired = stats.ttest_rel(a, b)
print(f"Cohen's d_z (paired) = {d_paired:.3f}")
print(f"t_paired = {t_paired:.3f}, p = {p_paired:.4f}")
```

```python
# Convert t to r (effect size from t-test)

def r_from_t(t_value, df):
    # absolute magnitude; keep sign if you want direction with group ordering
    return np.sqrt((t_value**2) / (t_value**2 + df))

# from the previous independent t-test:
df_ind = len(x) + len(y) - 2
r_eff = r_from_t(t_stat, df_ind)
print(f"Effect size r (from t) = {r_eff:.3f}")
```

```python
# η² and partial η² from ANOVA (statsmodels)

import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Simulated one-way ANOVA data
g1 = np.random.normal(50, 5, 30)
g2 = np.random.normal(52, 5, 30)
g3 = np.random.normal(56, 5, 30)
df_anova = pd.DataFrame({
    "y": np.r_[g1, g2, g3],
    "grp": ["A"]*30 + ["B"]*30 + ["C"]*30
})

model = ols("y ~ C(grp)", data=df_anova).fit()
anova_tbl = anova_lm(model, typ=2)
print(anova_tbl)

# η² for one-way (SS_between / SS_total)
ss_between = anova_tbl.loc["C(grp)", "sum_sq"]
ss_within  = anova_tbl.loc["Residual", "sum_sq"]
eta_sq = ss_between / (ss_between + ss_within)
print(f"η² (one-way) = {eta_sq:.3f}")

# For multi-factor ANOVA, compute partial η² per effect:
# ηp² = SS_effect / (SS_effect + SS_error)
# Here with one factor it's identical to η².
eta_p_sq = ss_between / (ss_between + ss_within)
print(f"Partial η² (one-way) = {eta_p_sq:.3f}")
```

```python
# Cramér’s V for a contingency table

import numpy as np
from scipy.stats import chi2_contingency

# 2x3 contingency example
table = np.array([[30, 21, 9],
                  [25, 15, 20]])

chi2, p, dof, exp = chi2_contingency(table, correction=False)
n = table.sum()
r, c = table.shape
cramers_v = np.sqrt(chi2 / (n * min(r-1, c-1)))

print(f"Chi² = {chi2:.3f}, df = {dof}, p = {p:.4f}")
print(f"Cramér's V = {cramers_v:.3f}")
```

```python
# Cohen’s h for two proportions

from statsmodels.stats.proportion import proportion_effectsize

# Example: conversions
x1, n1 = 200, 2000  # 10%
x2, n2 = 273, 2100  # 13%
p1, p2 = x1/n1, x2/n2

# Cohen's h
h = proportion_effectsize(p1, p2)
print(f"Cohen's h = {h:.3f}")

# For reference, also show raw difference
print(f"Risk difference = {p1 - p2:.3f}")
```

**Interpretation guides**

- **Cohen’s d** (means): small ≈ 0.2, medium ≈ 0.5, large ≈ 0.8  
- **r** (correlation / from t): small ≈ 0.10, medium ≈ 0.30, large ≈ 0.50  
- **η² / η²ₚ** (ANOVA): small ≈ 0.01, medium ≈ 0.06, large ≈ 0.14  
- **Cramér’s V** (categorical): small ≈ 0.10, medium ≈ 0.30, large ≈ 0.50  
- **Cohen’s h** (proportions): small ≈ 0.20, medium ≈ 0.50, large ≈ 0.80

## 23. Decision Map for Hypothesis Testing

This map helps us decide **which statistical test** to use based on:
- the **goal** (compare means, proportions, or associations),
- the **number of groups**, and
- whether **parametric assumptions** hold.



### 1. Quantitative (Continuous) Data - Comparing Means

| Situation | Assumptions | Recommended Test | Python function |
|------------|-------------|------------------|-----------------|
| **1 group** vs known $\mu_0$ | Population $\sigma$ known | One-sample **z-test** | `ztest()` |
| **1 group** vs known $\mu_0$ | $\sigma$ unknown, normal data | One-sample **t-test** | `stats.ttest_1samp()` |
| **2 independent groups** | Normal, equal variances | Student’s **t-test** | `stats.ttest_ind(equal_var=True)` |
| **2 independent groups** | Normal, unequal variances | Welch’s **t-test** | `stats.ttest_ind(equal_var=False)` |
| **2 paired samples** | Differences normal | **Paired t-test** | `stats.ttest_rel()` |
| **2 paired samples** | Non-normal differences | **Wilcoxon signed-rank** | `stats.wilcoxon()` |
| **>2 independent groups** | Normal, equal variances | **One-way ANOVA** | `anova_lm()` |
| **>2 independent groups** | Normal, unequal variances | **Welch’s ANOVA** | `anova_oneway(use_var='unequal')` |
| **>2 independent groups** | Non-normal | **Kruskal–Wallis** | `stats.kruskal()` |
| **>2 related measures** | Repeated measures, normal | **Repeated-measures ANOVA** | `AnovaRM()` |
| **>2 related measures** | Non-normal | **Friedman test** | `stats.friedmanchisquare()` |

---

### 2. Categorical Data - Comparing Counts or Proportions

| Situation | Recommended Test | Python function |
|------------|------------------|-----------------|
| **1 sample** (fit to theoretical distribution) | **Chi-square Goodness-of-Fit** | `stats.chisquare()` |
| **2 independent groups** (proportions) | **Two-proportion z-test (A/B test)** | `proportions_ztest()` |
| **>2 groups**, contingency table | **Chi-square Test of Independence** | `stats.chi2_contingency()` |
| **2×2 table, small expected counts** | **Fisher’s Exact Test** | `stats.fisher_exact()` |

---

### 3. Correlations & Associations

| Goal | Data Type | Recommended Test | Python function |
|------|------------|------------------|-----------------|
| Linear association | Continuous, normal | **Pearson correlation** | `stats.pearsonr()` |
| Monotonic association | Continuous, non-normal | **Spearman rank** | `stats.spearmanr()` |
| Ordinal association | Ordinal data | **Kendall’s tau** | `stats.kendalltau()` |

---

### 4. Assumption Checks (Diagnostics)

| Purpose | Test | Python function |
|----------|------|-----------------|
| Normality | Shapiro–Wilk / D’Agostino $K^2$ | `stats.shapiro()`, `stats.normaltest()` |
| Equal variances | Levene / Bartlett | `stats.levene()`, `stats.bartlett()` |
| Independence | Durbin–Watson, Runs test | `statsmodels.stats.stattools.durbin_watson()` |
| Outliers | Z-scores / IQR rule | — |

---

### 5. Effect Size & Power

| Goal | Metric | Formula / Function |
|------|---------|--------------------|
| Standardized mean difference | Cohen’s $d$ | $d = \frac{\bar x_1 - \bar x_2}{s_p}$ |
| Strength of association | Cramér’s $V$ | $V = \sqrt{\frac{\chi^2}{n(k-1)}}$ |
| Proportion difference | Cohen’s $h$ | $h = 2(\arcsin\sqrt{p_1}$ - $\arcsin\sqrt{p_2})$ |
| Variance explained | $\eta^2$, $\eta_p^2$ | `anova_lm()` output |
| Power analysis | Determine required $n$ | `TTestPower().solve_power()` |

---

### Decision flow summary

1. **Identify your dependent variable** (continuous or categorical).  
2. **Count your groups** (1, 2, or >2).  
3. **Check assumptions** (normality, equal variances).  
4. **Choose test accordingly** (parametric / nonparametric).  
5. **Report effect size and power** (practical significance).

---

**Example quick logic:**

> • Two groups, continuous variable → check normality & variances →  
> &nbsp;&nbsp;→ normal + equal → Student’s t  
> &nbsp;&nbsp;→ normal + unequal → Welch’s t  
> &nbsp;&nbsp;→ non-normal → Mann–Whitney U  
>
> • Three groups → ANOVA / Welch ANOVA / Kruskal–Wallis  
>
> • Categorical counts → Chi-square / Fisher  
>
> • Correlation → Pearson / Spearman / Kendall

```python
# p-value

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# parameters
alpha = 0.05
mu0, sigma = 0, 1
x = np.linspace(-4, 4, 400)
pdf = norm.pdf(x, mu0, sigma)
crit = norm.ppf(1 - alpha/2)

# observed statistic
t_obs = 2.1
p_value = 2 * (1 - norm.cdf(abs(t_obs)))

# plot
plt.figure(figsize=(8,4))
plt.plot(x, pdf, lw=2)
plt.fill_between(x, pdf, 0, where=(x <= -crit) | (x >= crit),
                 color="tomato", alpha=0.4,
                 label="Rejection region (α/2 each side)")
plt.axvline(t_obs, color="dodgerblue", lw=2,
            label=f"Observed eg. t = {t_obs:.2f}")
plt.axvline(-t_obs, color="dodgerblue", lw=1, ls="--", alpha=0.6)

# add p-value text box
plt.text(2.6, 0.05,
         f"a = 0.05 \np-value = {p_value:.4f} < a",
         fontsize=11,
         bbox=dict(facecolor="lightyellow",
                   edgecolor="gray",
                   boxstyle="round,pad=0.4"))

plt.title("p-value and critical regions example")
plt.xlabel("Test statistic")
plt.ylabel("Density under H₀ ~ N(0,1)")
plt.legend()
plt.tight_layout()

plt.savefig("figures/example-p-value-c-regions.png",dpi=160, facecolor="white", bbox_inches="tight")
plt.show()
```
